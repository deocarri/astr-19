```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def linspace():           #name of the definition
    x = np.linspace(0,1,100)   
    return np.exp(x)       #return the exp of x
```

```python
x = np.arange(0,1,0.01)    
y = linspace()                    #make y equal to the def
plt.plot(x,y)                     #plots x and y into a graph
plt.xlabel('Time [milliseconds]') #label x and y
plt.ylabel('Awesomeness')
plt.savefig('Session 7 Prompt.pdf',bbox_inches = 'tight',dpi=600)
```
