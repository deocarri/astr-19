```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#from 0 to 1 in 100 steps
def linspace():
    x = np.linspace(0,1,100)
    return x
```

```python
#define sine
def sine(x):
    return np.sin(x)
```

```python
#define cosine
def cos(x):
    return np.cos(x)
```

```python
#1 row, 2 columns of figures
f, axarr = plt.subplots(1,2)

#first array
axarr[0].plot(x, sine(linspace()))
axarr[0].set_xlabel('x')
axarr[0].set_ylabel('sin(x)')
axarr[0].set_title(r'$\sin(x)$')

#second panel
axarr[1].plot(x,cos(linspace()))
axarr[1].set_xlabel('x')
axarr[1].set_ylabel('cos(x)')
axarr[1].set_title(r'$\cos(x)$')

#save the figure into a pdf
plt.savefig("Session 8.pdf", bbox_inches='tight', dpi=600)
```
